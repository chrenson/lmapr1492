```python
from pymatgen.analysis.diffraction.core import AbstractDiffractionPatternCalculator
from pymatgen.analysis.diffraction.xrd import XRDCalculator
from pymatgen.ext.matproj import MPRester

with MPRester("6t2dQIcMaNaz6nTZlWm") as m:

    # Structure du matériau attribué
    structure = m.get_structure_by_material_id("mp-867339")

lambd = 1.54060
calculator = XRDCalculator(lambd) #Computes the XRD pattern of a crystal structure
pattern = calculator.get_pattern(structure) #Calculates the diffraction pattern for the structure
two_theta = pattern.x #two theta angles
intensities = pattern.y #Intensities
hkl = pattern.hkls 


```

# Figure de diffraction

```python
plot = calculator.show_plot(structure, annotate_peaks = None)
```

# Premier pic du diffractogramme 

```python
theta1 = two_theta[0]
intensity1 = intensities[0]
hkl1 = hkl[0]
print("Le premier pic du diffractogramme est situé à 2\u03B8 = " , theta1)
print("Son intensité est de ", intensity1," (sur une échelle de 100)")
print("Les indices hkl associés sont :\n", hkl1[0]['hkl'], "de multiplicité", hkl1[0]['multiplicity'])
print(hkl1[1]['hkl'], "de multiplicité", hkl1[1]['multiplicity'])
```
